## Plotting Complex Function

Consider a function 

$$f(z) = a + bz + cz^2$$ 

where $a, b, c \in \mathcal{R}$.

- Plot the real and imaginary parts of the function $f(z)$ where $z \in [-3,3]$.
- Plot the real and imaginary parts of the function $f(x+iy)$ where $x \in [-3,3]$ and $y = \{-1,0,1\}$.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solution-1
a=10*np.random.random()-5
b=10*np.random.random()-5
c=10*np.random.random()-5
a=np.random.random()
def f(z):
    return a+b*z+c*z**2
z = np.arange(-3,3,0.001)
func=f(z)
plt.plot(z,np.real(func),'r', label="real")
plt.plot(z,np.imag(func),'g--',label="imaginary")
plt.xlabel(r"$z$")
plt.ylabel(r"$f(z)$")
plt.legend()
plt.show()
```

```python
#solution->2
x=np.arange(-3,3,0.01)
y_values= [-1,0,1]
colors= ["red","green","blue"]
fig, ax = plt.subplots(1, 3, figsize=(15,5))
for y,color,axis in zip(y_values,colors,ax):
    axis.plot(x,np.real(f(x+1j*y)),color=color,linestyle='-',label="real",alpha=1.0)
    axis.plot(x,np.imag(f(x+1j*y)),color=color,linestyle='--',label="imaginary",alpha=1.0)
    axis.set_xlabel(r"$x$")
    axis.set_ylabel(r"$f(z)$")
    axis.legend()
    

plt.tight_layout()    
plt.show()
```
